```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(range(0, 30000, 100), columns=['episodes'])
df.head()
```

```python
df['a'] = 0.1
df['b'] = 0.9

df['c'] = (600 / 3000) * 0.9 + 0.1
df.loc[0,'c'] = 1

df['d'] = 0.05 + (1 - 0.05) * math.e ** (-df['episodes'] / 6000)

rate = 12/10
df['e'] = 0.9 * (0.1/0.9) ** (df['episodes'] ** rate / 6000 ** rate)
mask = df['e'] < 0.1
df.loc[mask, 'e'] = 0.1

rate = 12/10
df['f'] = 0.1 + (0.9-0.1) * (((2**rate)-2**(rate*df['episodes']/6000))/((2**rate)-1))
mask = df['f'] < 0.1
df.loc[mask, 'f'] = 0.1

df['g'] = 0.99
df.head()
```

```python
df.plot(x='episodes')
```

```python
plt.figure(layout='constrained')
plt.plot(df['episodes'], df['d'], label='d')
for e in 'abcdefg':
    plt.plot(df['episodes'], df[e], label=e)
plt.xlabel('episódios')
plt.ylabel('épsilon')
plt.legend(title='épsilon')
```
